### Introduction

International migration has become a major driver of demographic change in cities worldwide, reshaping urban landscapes and creating new demands for housing, transport, and social infrastructure (OECD, 2023). New arrivals often encounter barriers to accessing affordable housing, employment opportunities, language support, and essential public services, delaying social and economic integration (Coumans and Wark, 2024).

In Glasgow, these issues are particularly pronounced, exhibiting some of the highest levels of deprivation in the UK, alongside a rapidly changing urban landscape and ongoing regeneration programmes (Egan et al., 2015). This issue is exacerbated in Scotland, where housing policy ensures all those unitentionally homeless have the right to permenant housing, egendering Glagsow a hotspot for new asylum seekers despite ongoing housing crisis (Scottish Housing News, 2025). Glasgow currently houses the most aslyum seekers - following London - of any UK city, a number that will only continue to grow with the past year's record-high number of asylum claims (Cochrane, 2025).

To address these challenges, this study investigates the following research question:

Which areas of Glasgow are most suitable for asylum seeker housing based on a Multi-Criteria Evaluation (MCE) integrating socioeconomic, infrastructural, and environmental factors?

Ensuring safe housing infrastructure for asylum seekers is a critical step ascertaining social cohesion and integration of marginilsied groups within an existing soceity (Brown et al., 2022). Despite aslyum seekers provding a proven socioeconmic benefit to the UK (Together with Refugees, 2025), prominent framing of asylum seekers in the UK is largely negatively connoted, substantiated by practiced policies of deliberate destitution upon arrival. Growing populist rhetoric within the UK enables the labellign of aslyum seekers as "undeserving", a burden on public welfare and employment opportunites (Ramachandran, 2024). Despite Scotland's more forgiving devolved asylum seekers policies, anti-immigration narratives are surging (David Hume Institute, 2025).

UK aslyum housing, is thereby "embedded in racial capitalism", repoducing on the interesction of racialisation, classism and gendered exclusions, whereby Glasgow's aslyum housing stuggle has become a "key terrain" in exhibiting the issue (Santamarina and Karaliotas, 2025).

Chosing where to place future aslyum seekers, or devloping alsyum seeker housing projects, is contigent on a mulitiude of factors.

Previous resarch highligts migrant settlement patterns in areas with lower housing costs, better public transport access, and proximity to community services that provide social, cultural, and linguistic assistance (Bolt et al., 2010; Phillimore, 2020). Studies in spatial accessibility have also demonstrated that the distribution of public services can exacerbate inequalities when marginalised communities are located far from essential facilities (Maheshwari et al., 2024).

Concurrerently, preexisting environmental and infrastructural constriants - such as flood zones, steep terrain and land-use limitations coincide to subject availability of suitable devlopment land accross all ciies.

Spatial MCE is an established method for integrating heterogeneous datasets, including socioeconomic, environmental, and infrastructural indicators, to support land suitability analysis. **2.1 Data and Study Area**

The study focuses on the city of Glasgow, Scotland's largest urban centre and a major UK asylum dispersal location. Glasgow is characterised by pronounced socioeconomic inequalities, distinct neighbourhood-level variations in service accessibility, and diverse migrant communities. These characteristics make it an appropriate context for assessing spatial housing suitability for asylum seekers.

Multiple datasets were compiled to represent socioeconomic conditions, accessibility factors, and environmental constraints. These datasets were sourced from publicly available government and organisational repositories.

Each dataset was processed into raster, vector or CSV (for python) formats.
The main data layers used in this study include:
- Glasgow city outline boundary sourced from scottish census and produced into polygon to clip layers to.
- Community Centre Shapefile was soruced using Open Street Map quick query; installed as a plugin on QGIS. Field input was community centres and values inlcuded all froms of community centres. Data was clipped to glasgow boundaries
- Slope rasters were downlaoded from preexisting topographical data wihtin the author's repository.
- Schools and education centre shapefile was downloaded from gov.uk and clipped to glasgow boundary.
- Public Transport links was downloaded via National Public Transport Access Nodes (NAPTAN). Bus stops within the 'glasgow city' ward were filtered and downloaded as shapefiles.

Constraint layers were used to mask unsuitable land entirely or downweight suitability in high-risk or inacessible zones included
- Areas of steep slope; above 5 degrees gradient (derived from a DEM)
- Land use restrictions such as water bodies like the River Clyde were removed during the polygonisation of 'Glasgow City' boundaries.


All layers were projected into the British National Grid (EPSG:27700) and resampled to a consistent spatial resolution for integration within the MCE workflow.

Euclidean distance was used to convert each feature into a continuous raster.

**2.2 Multi-Criteria Evaluation Framework**

A MCE approach employing weighted factors of proximity to schools, proximity public transport nodes, proximity community centres and slope gradient, was used to assess suitability influencing suitability. MCE is appropriate for housing assessments because it allows continuous trade-offs between multiple criteria while remaining transparent and interpretable for planning purposes.

In order to acheive MCE:
1. Bus stop, schools and community centre shapefiles were converted to rasters.
2. Euclidean distance was used to convert each feature into a continuous raster.
3. Each factor was normalised to a 0-1 suitability scale to ensure comparability.
4. Slope constraint - existing as a raster was subsequently normalised.
5. Areas of slope above 5 degrees was turned into a constraint using raster calculator
6. Final suitability map of weighted factors was calculated.

Normalisation direction reflected the suitability logic. Proximity to public transport, healthcare, and schools was modelled as a “benefit” factor, where shorter distances correspond to higher suitability. Socioeconomic disadvantage was treated as a proxy for housing affordability; following previous research, areas with higher deprivation scores were considered more likely to contain low-cost housing suitable for asylum seeker accommodation. Migrant population density was included to reflect the importance of existing community networks in supporting integration.

**2.3 Weighted Factors**

To combine the different spatial criteria into a single Multi-Criteria Evaluation (MCE) surface, each input raster was assigned a weighting factor based on its relative importance to the analysis. The weights reflect the degree to which each variable contributes to the suitability outcome and were normalised so that more influential factors exert a proportionally greater effect on the final model.

Bus stop accessibility was assigned the highest weight among the service-based factors (0.4), as proximity to public transport is considered a key determinant of accessibility within the study area. Schools were weighted at 0.3, reflecting their importance while acknowledging they have a slightly lower influence relative to transport access. Community centres were assigned a weight of 0.2, as they provide relevant community support but are considered less essential in determining overall suitability compared to transport and education services.

In contrast to these service-based variables, slope was treated as a constraint rather than a standard criterion. Because steep terrain can significantly restrict development potential, slope was assigned the maximum weight of 1. This ensured that unsuitable terrain was strongly penalised in the final calculation, overriding otherwise favourable locations.

These weighted factors were then combined using a weighted linear combination to generate the final suitability surface.

### Calculating Final Suitability Map

Slope Constraint was calculated as follows:
("slope_norm@1" <= 0.05) * 1.
Any slope beyond 5 degrees was deemed not suitable for project devlopment site.

The following formula was inputted into raster calculator to create the final suitability map

(
  ("busstops_norm@1" * 0.4) +
  ("schools_norm@1" * 0.3) +
  ("community_norm@1" * 0.2) +
  ((1 - "slope_norm@1") * 0.1)
) * "slope_constraint@1"

## Discussion

The suitability map of housing for asylum seekers in Glasgow provides important insights into the spatial distribution of potential housing opportunities across the city, reflecting both the characteristics of the urban environment and the weighted criteria used in the analysis. Overall, the map is predominantly blue, indicating that many areas score highly in suitability, though clear spatial variations exist, highlighting areas of both advantage and constraint.

A key driver of high suitability across the city is transport accessibility, which received the highest weight in the multi-criteria evaluation. Glasgow benefits from a relatively well-developed public transport network, including buses, trains, and the subway, which provides frequent and reliable connections to essential services and employment. This finding is consistent with the literature on refugee and asylum seeker integration, which emphasizes that proximity to transport significantly enhances access to education, healthcare, and employment opportunities, improving social inclusion and reducing isolation (Ager and Strang, 2008). The widespread accessibility explains why most areas in Glasgow appear highly suitable and underscores the critical role of transport infrastructure in shaping housing suitability for vulnerable populations.

Another factor contributing to high suitability is the distribution of schools, which ensures that children of asylum seekers have access to education within a reasonable distance. This aligns with research emphasizing the importance of educational access for the integration and wellbeing of migrant children (Dryden-Peterson, 2016). By including schools as a key criterion, the analysis recognizes the dual importance of physical accessibility and social infrastructure in promoting positive settlement outcomes.

The southeast of Glasgow consistently appears as one of the least suitable areas in the suitability map, and this pattern is strongly supported by existing research on urban spatial inequality. Studies in UK cities have shown that peripheral neighbourhoods often experience reduced access to public transport, fewer public services, and lower levels of institutional investment (Pacione, 2004; Lucas et al., 2019). These dynamics are clearly reflected in the southeast of Glasgow, where transport connections are comparatively weaker and educational facilities and other public services are more sparsely distributed.

Because transport accessibility carried the highest weighting in the MCE, areas lacking frequent and reliable connections were heavily penalised. This aligns with Ager & Strang’s (2008) framework of integration, which positions access to services—especially mobility and education—as central indicators of a successful settlement environment for asylum seekers. Without strong transport links, asylum seekers face barriers to healthcare, social support, and employment (Fitzharris et al., 2023), making these southeastern neighbourhoods less conducive to successful integration.

To complement the MCE suitability analysis, vacancy rates across Glasgow were explored using a Python-based geopandas interactive map. The visualisation revealed that higher vacancy levels are clustered primarily in the South Central and Southwestern wards, including areas around Govan, Southside Central and Calton, which align with publicly reported ward-level vacancy patterns. These concentrations were clearly visible in the map, where darker-shaded polygons indicated a greater number or percentage of empty dwellings. In contrast, northwestern areas such as Maryhill and the West End displayed comparatively low vacancy, consistent with their more stable private rental and owner-occupied housing markets.

Although vacancy alone does not determine suitability the spatial overlap between high-vacancy zones and moderately suitable MCE areas suggests that South Central Glasgow may offer an initial focus point for pilot housing redevelopment or repurposing initiatives.

Overall, the divide between the unsuitable southeast and highly suitable northwest reflects long-standing patterns of urban spatial inequality, transport accessibility, and uneven service distribution, all of which are deeply documented in UK urban and migration literature. The suitability map therefore not only reflects your criteria weights, but also echoes broader structural patterns affecting asylum seeker integration in Glasgow.


## Limitations

This study was restricted by limited data access, meaning only a small number of suitable datasets - mainly transport, schools, slope and community facilities - could be included in the model. Important factors such as healthcare, crime, housing quality, and deprivation were unavailable at the required spatial scale, which simplified the analysis and influenced the weighting.

Beyond data availability, several broader limitations affected this study. The multi-criteria evaluation relies heavily on subjective weighting, and although weights were informed by the literature, different weighting choices could produce different outcomes. The model also assumes that all criteria operate independently, which oversimplifies the complex social and spatial factors influencing asylum seeker settlement.

## Conclusion

The analysis shows that most of Glasgow has high suitability for potential future asylum seeker housing projects, particularly the northwest and central areas, due to strong transport links and good access to schools. In contrast, the southeast is consistently less suitable, reflecting broader patterns of limited service provision and weaker connectivity. These findings highlight both opportunities and gaps within Glasgow’s urban landscape. The project contributes, spatial assessment of suitability and demonstrates how multi-criteria evaluation can support evidence-based planning for asylum seeker accommodation, even with limited datasets.
